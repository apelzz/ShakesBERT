```python
import json
import matplotlib
import matplotlib.pyplot as plt

with open("MM_acc_cos_log_lr5e-7_100epochs.json", "r") as openfile:
    MM_json = json.load(openfile)
    
with open("MMMasked_acc_cos_log_lr5e-7_100epochs.json", "r") as openfile:
    MMMasked_json = json.load(openfile)

with open("Base_acc_cos_log_lr5e-7_100epochs.json", "r") as openfile:
    Base_json = json.load(openfile)
    
MM_arr = MM_json["acc_arr"]
MMMasked_arr = MMMasked_json["acc_arr"]
Base_arr = Base_json["acc_arr"]

MM_cos = MM_json['cos_arr']
MMMasked_cos = MMMasked_json["cos_arr"]
Base_cos = Base_json['cos_arr']


plt.plot(MM_arr, '-o', label = 'MM unmasked accuracies', color = 'blue')
plt.plot(MMMasked_arr, '-o', label = 'MM masked accuracies', color = 'green')
plt.plot(Base_arr, '-o', label = 'Base accuracies', color = 'orange')
plt.xlabel('Checkpoint Number')
plt.ylabel('Top-5 test accuracy at checkpoint')
plt.legend()
plt.show()

plt.plot(MM_cos, '-o', label = 'MM unmasked cosine similarity', color = 'blue')
plt.plot(MMMasked_cos, '-o', label = 'MM masked cosine similarity', color = 'green')
plt.plot(Base_cos, '-o', label = 'Base cosine similarity', color = 'orange')
plt.xlabel('Checkpoint Number')
plt.ylabel('Top-5 test accuracy at checkpoint')
plt.legend()
plt.show()

```
