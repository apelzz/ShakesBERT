```python
# Convert shakestrain prosodic output to syllables

input_dir = "./train_data_prosodic/"
# input_dir = "./train_data_prosodic/sonnets_train_split/"
output_dir = "./train_data_syllables/"
# output_dir = "./train_data_syllables/sonnets_train_split/"

data_type = "shakestrain"

input_file_name = f"{input_dir}{data_type}_train_prosodic.txt"
output_file_name = f"{output_dir}{data_type}_train_syllables.txt"

# Read input file 
input_file = open(input_file_name, "r", encoding="utf-8")
input_text = input_file.read()

# Split input text into a 2d list of words 
input_text = input_text.split("\n")
temp = []
for line in input_text:
    temp.append(line.split("|"))    
input_text = temp

# Append stress to each syllable
for i in range(len(input_text)):
    for j in range(len(input_text[i])):
        if input_text[i][j].islower():
            input_text[i][j] += "0"
        else: 
            input_text[i][j] += "1"
# input_text.pop() # Remove the last empty line

# Join words back into a string
temp = []
for line in input_text:
    temp.append(' '.join(line))
input_text = temp
input_text = '\n'.join(input_text)

# print(input_text)
output_file = open(output_file_name, "w", encoding="utf-8")
output_file.write(input_text)

input_file.close()
output_file.close()
```

```python
# Convert sonnets prosodic output to syllables

# input_dir = "./train_data_prosodic/"
input_dir = "./train_data_prosodic/sonnets_train_split/"
# output_dir = "./train_data_syllables/"
output_dir = "./train_data_syllables/sonnets_train_split/"

data_type = "sonnets"

for file_num in range(42):
    input_file_name = f"{input_dir}{data_type}_train_prosodic_{file_num+1}.txt"
    output_file_name = f"{output_dir}{data_type}_train_syllables_{file_num+1:02d}.txt"

    # Read input file 
    input_file = open(input_file_name, "r", encoding="utf-8")
    input_text = input_file.read()

    # Split input text into a 2d list of words 
    input_text = input_text.split("\n")
    temp = []
    for line in input_text:
        temp.append(line.split("|"))    
    input_text = temp

    # Append stress to each syllable
    for i in range(len(input_text)):
        for j in range(len(input_text[i])):
            if input_text[i][j].islower():
                input_text[i][j] += "0"
            else: 
                input_text[i][j] += "1"
#     input_text.pop() # Remove the last empty line

    # Join words back into a string
    temp = []
    for line in input_text:
        temp.append(' '.join(line))
    input_text = temp
    input_text = '\n'.join(input_text)

    # print(input_text)
    output_file = open(output_file_name, "w", encoding="utf-8")
    output_file.write(input_text)

    input_file.close()
    output_file.close()
```

```python
# Combine 1k-line-sized sonnets syllables files into one large file

import glob

input_dir = "./train_data_syllables/sonnets_train_split/"

output_file_name ="./train_data_syllables/sonnets_train_syllables.txt"

read_files = glob.glob(f"{input_dir}*.txt")

with open(output_file_name, "wb") as outfile:
    for f in read_files:
        with open(f,"rb") as infile:
            outfile.write(infile.read())
```
